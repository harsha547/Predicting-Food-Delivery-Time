### Data Import

```python
test_filename = 'Data_Test.xlsx'
train_filename = 'Data_Train.xlsx'
Sample_Submission_filename = 'Sample_Submission.xlsx'

df_train = pd.read_excel(train_filename)
df_train.head()
```

```python
df_location_dummies = pd.get_dummies(df_train.Location)
```

```python
df_location_dummies.head()
```

```python
df_train['Restaurant'] = df_train['Restaurant'].str.replace('ID_', '')
df_train['Delivery_Time'] = df_train['Delivery_Time'].str.replace(' minutes', '')
df_train['Average_Cost'] = df_train['Average_Cost'].str.replace('₹', '')
df_train['Minimum_Order'] = df_train['Minimum_Order'].str.replace('₹', '')

df_train['Cuisines'] = df_train['Cuisines'] + ','
df_train['Cuisines'] = df_train['Cuisines'].str.split(',', n = 1, expand = True)[0]

df_train.to_csv('harsha.csv')
df_train
```

```python
df_cuisines_dummies = pd.get_dummies(df_train['Cuisines'])

df_cuisines_dummies.head()
```

```python
df_train.drop(['Location', 'Cuisines'], axis = 'columns', inplace=True)
df_cuisines_dummies.drop(['Awadhi'], axis = 1, inplace=True)
df_location_dummies.drop(['Delhi Cantt.'], axis = 1, inplace=True)
```

```python
df_merge = pd.concat([df_train, df_cuisines_dummies, df_location_dummies], axis='columns')
```

```python
df_merge.head()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
df_merge['Rating'] = df_merge['Rating'].replace({'-':0, 'NEW':0, 'Opening Soon':0, 'Temporarily Closed':0})
df_merge['Votes'] = df_merge['Votes'].replace({'-':0, 'NEW':0})
df_merge['Reviews'] = df_merge['Reviews'].replace({'-':0, 'NEW':0})
df_merge['Average_Cost'] = df_merge['Average_Cost'].replace({'for':0})
df_merge["Average_Cost"] = df_merge["Average_Cost"].str.replace(",","")

#df_merge.fillna(df_merge.mean(), inplace=True)
df_merge.to_csv('harsha.csv')
df_merge.dropna(inplace=True)
```

```python
y = df_merge['Delivery_Time']
X = df_merge.drop(['Delivery_Time'], axis='columns')
```

```python
X.info()
X.describe()
model.fit(X, y)

#filtered_df = X[X['Average_Cost'].isnull()]
#filtered_df
```
